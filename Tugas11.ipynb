```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing the Iris dataset with pandas
dataset = pd.read_csv('https://raw.githubusercontent.com/lauracantik/mathtools/main/healthcare-dataset-stroke-data.csv');

# Iloc function
x = dataset.iloc[:, [2, 3, 8]].values

pd.DataFrame(x)
```

```python
# Mencari nilai optimum dari tiap cluster untuk klasifikasi k-means
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
# Plotting the results onto a line graph, allowing us to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```

```python
# Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters = 2, init = 'k-means++', max_iter = 300, n_init = 100, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
#Visualising the clusters
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 2], s = 5, c = 'black', label = '2')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 2], s = 10, c = 'orange', label = '0')
```

```python
# Menampilkan centroids 
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'red', label = 'Centroids')

plt.legend()
```
